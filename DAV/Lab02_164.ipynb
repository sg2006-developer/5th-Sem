```python
#Q1
def fillwithmean(df,col):
    df[col] = df[col].fillna(df[col].mean())
    df['City'] = df['City'].fillna('Unknown')
    return df

df = fillwithmean(df,'Age')
```

```python
# Q2
def dropduplicateonCol(df):
    df=df.drop_duplicates()
    return df
```

```python
df = dropduplicateonCol(df)
```

```python
# Q3
def handle_inconsitent(df):
    mpp = {'M':"Male",'F':"Female"}
    df["Gender"] = df["Gender"].replace(mpp)
    return df
df = handle_inconsitent(df)
```

```python
def city_to_dummies(df):
    if 'City' in df.columns:
        dummies = pd.get_dummies(df['City'],prefix='City')
        df=pd.concat([df,dummies],axis=1)
    return df
df=city_to_dummies(df)
print("\n Dataframe: ")
print(df)
```
